```python
import pandas as pd

# 1) GitHub'daki CSV'yi oku
url = "https://raw.githubusercontent.com/guzelTemli/YapayZekaChatbot/refs/heads/main/bilgi_guvenligi_dataset/bilgi_guvenligi_chunks_v3_final.csv"
df = pd.read_csv(url)

print("Satır sayısı:", len(df))
df.head()
```

```python
!pip install deep-translator
from deep_translator import GoogleTranslator
import time

translator = GoogleTranslator(source="en", target="tr")

def translate_en_to_tr_safe(text):
    text = str(text)
    if not text.strip():
        return ""
    # Çok uzun metinler için Google bazen hata verebilir, bu yüzden try/except
    try:
        tr = translator.translate(text)
        # Baş harfi büyük yap, gereksiz boşlukları temizle
        return tr.strip()
    except Exception as e:
        print("Çeviri hatası, metin kısaltılıyor:", e)
        # Hata alırsak biraz bekleyip metnin ilk 4000 karakterini deneyelim
        time.sleep(1)
        try:
            tr = translator.translate(text[:4000])
            return tr.strip()
        except:
            return ""

# 2) Yeni sütun ekle: chunk_text_tr_clean
df["chunk_text_tr_clean"] = df["chunk_text"].apply(translate_en_to_tr_safe)

# 3) Yeni CSV olarak kaydet
output_path = "bilgi_guvenligi_chunks_v3_tr_clean.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")

output_path
```
